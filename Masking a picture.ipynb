```python
import cv2 
import matplotlib.pyplot as plt
import numpy as np
```

```python
comp=cv2.imread("computeronic.jpg",1)
gray_comp=cv2.cvtColor(comp,cv2.COLOR_BGR2GRAY)
plt.imshow(gray_comp,cmap="gray")
```

```python
back=cv2.imread("colorful.png",1)
back_rgb=back[:,:,::-1]
plt.imshow(back_rgb)
```

```python
back_rgb=cv2.resize(back_rgb,dsize=(1033,720),interpolation=cv2.INTER_AREA)
plt.imshow(back_rgb)
```

```python
retval,mask_comp=cv2.threshold(gray_comp,100,255,cv2.THRESH_BINARY_INV)
plt.imshow(mask_comp,cmap="gray")
```

```python
retval,mask_comp_1=cv2.threshold(gray_comp,100,255,cv2.THRESH_BINARY)
plt.imshow(mask_comp_1,cmap="gray")
```

```python
comp_selected=cv2.bitwise_and(comp,comp,mask=mask_comp)
plt.imshow(comp_selected,cmap="gray")
```

```python
back_rgb_selected=cv2.bitwise_and(back_rgb,back_rgb,mask=mask_comp_1)
plt.imshow(back_rgb_selected,cmap="gray")
```
